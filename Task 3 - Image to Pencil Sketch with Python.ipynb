<h1>Import Libraries</h1>

```python
import cv2
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
```

<h1>Loading and Plotting the Original image</h1>

```python
image = cv2.imread('C:\\Users\\DELL\\Downloads\\krishna.JPG')
plt.figure(figsize=(9,15))
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis("off")
plt.title("REAL IMAGE")
plt.show()
```

<h1>Converting Original image to Grayscale</h1>

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(9,15))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("GRAYSCALE IMAGE")
plt.show()
```

<h1>Inverting the Image</h1>

```python
image_invert=cv2.bitwise_not(image_gray)
plt.figure(figsize=(9,15))
plt.imshow(image_gray,cmap="gray")
plt.axis("off")
plt.title("INVERTED IMAGE")
plt.show()
```

<h1>Smoothing the Image</h1>

```python
blurred=cv2.GaussianBlur(image_invert,(21,21),0)
plt.figure(figsize=(9,15))
plt.imshow(blurred,cmap="gray")
plt.axis("off")
plt.title("BLURRED IMAGE")
plt.show()
```

<h1>Converting the image into pencil sketch</h1>

```python
pencil_sketch=cv2.divide(image_gray,255-blurred,scale=255)
plt.figure(figsize=(9,15))
plt.imshow(pencil_sketch,cmap="gray")
plt.title("FINAL IMAGE")
plt.axis("off")
```
